#### Q5

```python
#sampling
np.random.seed(4)
dev=np.random.rand(len(df))<=0.7
```

```python
big=df[dev]
```

```python
small=df[~dev]
```

```python
#hourly prices
hourly_prices = small.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
line= sns.lineplot(data=hourly_prices, x='order_hour_of_day', y='prices')
```

```python
hourly_prices_1= big.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
line_1= sns.lineplot(data=hourly_prices_1, x='order_hour_of_day', y='prices')
```
