```python
#limiting the database
df = df_ords_prods_merged[:1000000]
```

```python
#define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

#### to find most,fewest and average period for orders I'll create three period of times in a day that each contains 8 hours. (0-8, 8-16 and 16-24)

#### by looking the results I can say busiest period is 8-16 and least busiest period is 0-8

```python
results_periods = []
for value in df_ords_prods_merged['order_hour_of_day']:
    if value in (9, 10, 11, 12, 13, 14, 15, 16):
        results_periods.append('Most orders')
    elif value in (7, 8, 17, 18, 19, 20, 21, 22):
        results_periods.append('Average orders')
    else:
        results_periods.append('Fewest orders')
```

```python
df_ords_prods_merged['busiest_period_of_day']=results_periods
```
