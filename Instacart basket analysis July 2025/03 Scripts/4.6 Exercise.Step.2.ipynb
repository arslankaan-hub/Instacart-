```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# Import databases
path = r'C:\Users\arsla\Instacart basket analysis July 2025\02 Data'
```

```python
df_ords_prods_combined = pd.read_pickle(os.path.join(path,'Prepared Data','orders_products_combined.pkl'))
```

```python
df_prods = pd.read_csv(os.path.join(path,'Prepared Data','products_checked.csv'))
```

```python
#checking the databases
df_ords_prods_combined.shape
```

```python
df_ords_prods_combined.head()
```

```python
df_prods.head()
```

```python
#deleting unnamed columns
df_prods = df_prods.loc[:, ~df_prods.columns.str.contains('^Unnamed')]
```

```python
# merging databases
df_ords_prods_merged = df_ords_prods_combined.merge(df_prods, on = 'product_id', indicator = 'match')
```

```python
df_ords_prods_merged.head()
```

```python
#dropping unnecessary merge column
df_ords_prods_merged.drop('_merge', axis=1)
```

```python
#exporting dataframe as pickle
df_ords_prods_merged.to_pickle(os.path.join(path, 'Prepared data' ,'ords_prods_merge.pkl'))
```
